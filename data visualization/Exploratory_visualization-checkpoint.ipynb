## Univariate Exploration

````python
def plot_pie(df, col):
    plt.pie(loan_df[col].value_counts(), autopct='%.2f%%');
    plt.title(f'Proportion of {col}', fontsize=14)
    plt.legend(loan_df[col].value_counts().index, loc = 'upper right');

def plot_count(df, col, base_color=None, rotation=0, order=None, hue=None, palette=None, title=''):
        sns.countplot(data=df,x= col, color=base_color, order=order, hue=hue, palette=palette)
        plt.title(f'{title}')
        plt.xlabel(f'{col}')
        plt.ylabel('Count')
        plt.xticks(rotation=rotation);
````
